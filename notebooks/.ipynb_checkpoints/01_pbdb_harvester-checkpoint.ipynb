# 01 — PBDB Harvester
### Project IceWave | Pleistocene Megafauna | WA · OR · NV


```python
import requests, time, json
import pandas as pd
import geopandas as gpd
from pathlib import Path
from shapely.geometry import Point

Path('../data/pbdb').mkdir(parents=True, exist_ok=True)

TAXA = [
    'Mammuthus',
    'Mammut',
    'Megalonyx',
    'Nothrotheriops',
    'Equus',
    'Camelops',
    'Paramylodon',
]

STATES = {
    'WA': {'lngmin': -124.8, 'lngmax': -116.9, 'latmin': 45.5, 'latmax': 49.1},
    'OR': {'lngmin': -124.6, 'lngmax': -116.5, 'latmin': 41.9, 'latmax': 46.3},
    'NV': {'lngmin': -120.0, 'lngmax': -114.0, 'latmin': 35.0, 'latmax': 42.1},
}

BASE_URL = 'https://paleobiodb.org/data1.2/occs/list.json'

def fetch_with_retry(params, max_retries=4, timeout=90):
    """Fetch PBDB with retry and backoff."""
    for attempt in range(max_retries):
        try:
            r = requests.get(BASE_URL, params=params, timeout=timeout)
            if r.status_code == 200:
                return r.json().get('records', [])
            print(f'    HTTP {r.status_code} — retrying...')
        except requests.exceptions.Timeout:
            wait = 10 * (attempt + 1)
            print(f'    Timeout (attempt {attempt+1}/{max_retries}) — waiting {wait}s...')
            time.sleep(wait)
        except Exception as e:
            print(f'    Error: {e}')
            time.sleep(10)
    print(f'    FAILED after {max_retries} attempts')
    return []

all_records = []

for taxon in TAXA:
    for state, bbox in STATES.items():
        params = {
            'base_name': taxon,
            'interval':  'Pleistocene',
            'lngmin':    bbox['lngmin'],
            'lngmax':    bbox['lngmax'],
            'latmin':    bbox['latmin'],
            'latmax':    bbox['latmax'],
            'show':      'coords,loc,time,strat',
            'limit':     'all',
        }
        records = fetch_with_retry(params)
        for rec in records:
            rec['_taxon_query'] = taxon
            rec['_state']       = state
        all_records.extend(records)
        print(f'  {taxon:20s} {state}: {len(records):4d} occurrences')
        time.sleep(1)  # be polite to the API

print(f'\nTotal raw records: {len(all_records)}')
```

```python
# ── Parse into DataFrame ──────────────────────────────────────────────────
rows = []
for r in all_records:
    try:
        lat = float(r.get('lat', 0))
        lng = float(r.get('lng', 0))
    except:
        continue
    if lat == 0 and lng == 0:
        continue
    rows.append({
        'occurrence_no':  r.get('oid', ''),
        'taxon_name':     r.get('tna', r.get('genus_name', '')),
        'taxon_query':    r.get('_taxon_query', ''),
        'state':          r.get('_state', ''),
        'latitude':       lat,
        'longitude':      lng,
        'formation':      r.get('formation', ''),
        'stratgroup':     r.get('stratgroup', ''),
        'early_interval': r.get('early_interval', ''),
        'late_interval':  r.get('late_interval', ''),
        'collection_no':  r.get('cid', ''),
    })

df = pd.DataFrame(rows)
df = df.drop_duplicates(subset=['occurrence_no'])

print(f'Unique occurrences: {len(df)}')
print(f'\nBy taxon:')
print(df['taxon_query'].value_counts().to_string())
print(f'\nBy state:')
print(df['state'].value_counts().to_string())
```

```python
# ── Save CSV + GeoJSON ────────────────────────────────────────────────────
df.to_csv('../data/pbdb/icewave_occurrences.csv', index=False)

gdf = gpd.GeoDataFrame(
    df,
    geometry=[Point(row.longitude, row.latitude) for _, row in df.iterrows()],
    crs='EPSG:4326'
)
gdf.to_file('../data/pbdb/icewave_occurrences.geojson', driver='GeoJSON')

print(f'Saved {len(df)} records')
print(gdf[['taxon_name','state','latitude','longitude','formation']].head(10).to_string())
```

```python
# ── Summary map ───────────────────────────────────────────────────────────
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

colors = {
    'Mammuthus':      '#f0c040',
    'Mammut':         '#52b788',
    'Megalonyx':      '#74c0fc',
    'Nothrotheriops': '#ff8fab',
    'Equus':          '#ffa94d',
    'Camelops':       '#da77f2',
    'Paramylodon':    '#63e6be',
}

fig, axes = plt.subplots(1, 3, figsize=(18, 7))
fig.patch.set_facecolor('#1b4332')

states_info = [
    ('WA', -124.8, -116.9, 45.5, 49.1),
    ('OR', -124.6, -116.5, 41.9, 46.3),
    ('NV', -120.0, -114.0, 35.0, 42.1),
]

for ax, (state, lngmin, lngmax, latmin, latmax) in zip(axes, states_info):
    ax.set_facecolor('#0a1a0f')
    sub = df[df['state'] == state]
    for taxon, grp in sub.groupby('taxon_query'):
        ax.scatter(grp['longitude'], grp['latitude'],
                   c=colors.get(taxon, '#ffffff'), s=50, alpha=0.85,
                   label=taxon, edgecolors='none', zorder=3)
    ax.set_xlim(lngmin, lngmax)
    ax.set_ylim(latmin, latmax)
    ax.set_title(f'{state} — {len(sub)} occurrences',
                 color='#f0e8d0', fontsize=13, fontweight='bold')
    ax.tick_params(colors='#8b9bb4')
    for spine in ax.spines.values():
        spine.set_edgecolor('#2d6a4f')
    ax.grid(True, color='#2d6a4f', linewidth=0.4, alpha=0.5)

patches = [mpatches.Patch(color=c, label=t) for t, c in colors.items()]
fig.legend(handles=patches, loc='lower center', ncol=4,
           facecolor='#1b4332', edgecolor='#c9a84c',
           labelcolor='#f0e8d0', fontsize=10, bbox_to_anchor=(0.5, -0.02))
plt.suptitle('Project IceWave — PBDB Occurrences by State & Taxon',
             color='#c9a84c', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('../data/pbdb/occurrence_map.png', dpi=150,
            bbox_inches='tight', facecolor='#1b4332')
plt.show()
print('Map saved.')
```

```python
# ── Extra taxa + wider time interval ──────────────────────────────────────
# Catches specimens tagged as 'Late Pleistocene' or 'Rancholabrean' 
# rather than strictly 'Pleistocene'

EXTRA_TAXA = [
    'Mammuthus primigenius',   # woolly mammoth explicit
    'Mammuthus columbi',       # Columbian mammoth explicit  
    'Bison',                   # Pleistocene bison — good terrain proxy
    'Cervus',                  # elk — similar habitat signal
    'Arctodus',                # short-faced bear
]

INTERVALS = ['Pleistocene', 'Late Pleistocene', 'Rancholabrean']

extra_records = []

for taxon in EXTRA_TAXA:
    for interval in INTERVALS:
        for state, bbox in STATES.items():
            params = {
                'base_name': taxon,
                'interval':  interval,
                'lngmin':    bbox['lngmin'],
                'lngmax':    bbox['lngmax'],
                'latmin':    bbox['latmin'],
                'latmax':    bbox['latmax'],
                'show':      'coords,loc,time,strat',
                'limit':     'all',
            }
            records = fetch_with_retry(params)
            for rec in records:
                rec['_taxon_query'] = taxon
                rec['_state']       = state
            extra_records.extend(records)
            if records:
                print(f'  {taxon:30s} {interval:20s} {state}: {len(records):4d}')
            time.sleep(0.8)

print(f'\nExtra raw records: {len(extra_records)}')

# ── Merge with original ────────────────────────────────────────────────────
extra_rows = []
for r in extra_records:
    try:
        lat = float(r.get('lat', 0))
        lng = float(r.get('lng', 0))
    except:
        continue
    if lat == 0 and lng == 0:
        continue
    extra_rows.append({
        'occurrence_no':  r.get('oid', ''),
        'taxon_name':     r.get('tna', r.get('genus_name', '')),
        'taxon_query':    r.get('_taxon_query', ''),
        'state':          r.get('_state', ''),
        'latitude':       lat,
        'longitude':      lng,
        'formation':      r.get('formation', ''),
        'stratgroup':     r.get('stratgroup', ''),
        'early_interval': r.get('early_interval', ''),
        'late_interval':  r.get('late_interval', ''),
        'collection_no':  r.get('cid', ''),
    })

df_extra = pd.DataFrame(extra_rows)
df_combined = pd.concat([df, df_extra], ignore_index=True)
df_combined = df_combined.drop_duplicates(subset=['occurrence_no'])

print(f'\nCombined unique occurrences: {len(df_combined)}')
print(f'  Original:  {len(df)}')
print(f'  Added:     {len(df_combined) - len(df)}')
print(f'\nBy taxon:')
print(df_combined['taxon_query'].value_counts().to_string())
print(f'\nBy state:')
print(df_combined['state'].value_counts().to_string())
```

```python
# ── Save combined dataset ─────────────────────────────────────────────────
df_combined.to_csv('../data/pbdb/icewave_occurrences.csv', index=False)

gdf_combined = gpd.GeoDataFrame(
    df_combined,
    geometry=[Point(row.longitude, row.latitude) for _, row in df_combined.iterrows()],
    crs='EPSG:4326'
)
gdf_combined.to_file('../data/pbdb/icewave_occurrences.geojson', driver='GeoJSON')

# Overwrite df for use in map cell
df = df_combined
print(f'Saved {len(df_combined)} records to CSV + GeoJSON')
```
